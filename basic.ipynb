np.random.default_rng を使ってシード固定の 5x5 乱数行列を作る。

```python
rng1=np.random.default_rng(1)
rndA=rng1.random([5,5])
rndA
```

```python
x1=rng1.integers(10,size=6)
print(x1)
print(x1[:3])
print(x1[2:4])
print(x1[3:])
print(x1[::2])
```

```python
x2=rng1.integers(10,size=[5,5])
print(x2)
print(x2[:3,:3])
print(x2[::-1,::-1])
print(x2.T[::-1,::])
```

スライスがビューであることを確認し、コピーとの違いを実験で示す。

```python
y1=rng1.integers(10,size=6)
yy1=y1[:3]
yy1[0]=5
print(yy1,id(yy1[0]))
print(y1,id(y1[0]))
```

```python
y1=rng1.integers(10,size=6)
yy1=y1[:3].copy()
yy1[0]=5
print(yy1,id(yy1[0]))
print(y1,id(y1[0]))
```

reshapeで 3x4 -> 2x6 -> 12 に変換し、元配列と同じメモリを共有しているか調べる。

```python
B=rng1.integers(10,size=[3,4])
print(B,id(B[0][0]))
C=B.reshape([4,3])
C[0][0]=100
print(B,id(B[0][0]))
print(C,id(C[0][0]))
```

加減乗除・累乗・剰余のベクトル化計算を示す。

```python
x = np.arange(4)
print("x =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x // 2 =", x // 2) # 整数除算
```

```python
print(np.sin(x*np.pi/4))
print(np.exp(x))
print(np.log1p(x))
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(1, 10).reshape(3, 3), columns=['A', 'B', 'C'])
print(df)
#    A  B  C
# 0  1  2  3
# 1  4  5  6
# 2  7  8  9

# 軸なし: データ全体の合計
print(df.sum())  # 列ごとの合計が出力される
print(df.values.sum()) # numpyの配列に変換すると全体合計が取得できる
print(np.sum(df)) # 同じく、全体合計が取得できる
```

2D配列でaxis=0/1の合計を比較する。

```python
print(df.sum(axis=0))  # 各列の合計
# A    12
# B    15
# C    18
# dtype: int64

print(df.mean(axis=0)) # 各列の平均
# A    4.0
# B    5.0
# C    6.0
# dtype: float64
```

```python
print(df.sum(axis=1))  # 各行の合計
# 0     6
# 1    15
# 2    24
# dtype: int64

print(df.max(axis=1)) # 各行の最大値
# 0    3
# 1    6
# 2    9
# dtype: int64
```
