# Matplotlib

線グラフ

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-3,3,100)
y=np.sin(x)
plt.title('sin(x)')
plt.ylabel('sin(x)')
plt.xlabel('x')
plt.plot(y)
plt.show()
```

```python
rng=np.random.default_rng(1)
x=rng.integers(0,100,size=100)
y=rng.integers(0,100,size=100)
xy=zip(x,y)
for x,y in xy:
    if x+y>=100:
        
        plt.scatter(x,y,marker='o',color='red')#markerで渡さないとエラーになる
    else:
        plt.scatter(x,y,marker='.',color='blue')
plt.show()
```

```python
random_numbers = np.random.normal(10, 2, 1000)
#random_numbers
plt.hist(random_numbers,bins=10,alpha=0.5,color='red')
plt.hist(random_numbers,bins=100,alpha=0.5)
plt.show()
```

```python
plt.errorbar(random_numbers,range(len(random_numbers)))
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30)
cb = plt.colorbar()
cb.set_label('counts in bin')
```
