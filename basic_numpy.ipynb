```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(1, 10).reshape(3, 3), columns=['A', 'B', 'C'])
print(df)
#    A  B  C
# 0  1  2  3
# 1  4  5  6
# 2  7  8  9

# 軸なし: データ全体の合計
print(df.sum())  # 列ごとの合計が出力される
print(df.values.sum()) # numpyの配列に変換すると全体合計が取得できる
print(np.sum(df)) # 同じく、全体合計が取得できる
```

2D配列でaxis=0/1の合計を比較する。

```python
print(df.sum(axis=0))  # 各列の合計
# A    12
# B    15
# C    18
# dtype: int64

print(df.mean(axis=0)) # 各列の平均
# A    4.0
# B    5.0
# C    6.0
# dtype: float64
```

```python
print(df.sum(axis=1))  # 各行の合計
# 0     6
# 1    15
# 2    24
# dtype: int64

print(df.max(axis=1)) # 各行の最大値
# 0    3
# 1    6
# 2    9
# dtype: int64
```

np.sortとnp.argsortの違いを確認し、上位k件を取り出す。

```python
rng3=np.random.default_rng(3)
a1=rng3.integers(0,100,size=100)
c1=a1.copy()
k=3
#%timeit np.sort(a1)
#%timeit np.argsort(a1)
s1=np.sort(c1)
print(s1)
as1=np.argsort(c1)
print(s1[:k])
print(as1)
x=c1[as1]
print(x)
```
